# Kernel PCA
Kernel PCA is a non-linear version of PCA. It uses kernel functions to project the data into a higher-dimensional space where linear separation may be more easily achieved. It's particularly useful for datasets where the relationship between features is non-linear.


**Purpose:** Captures non-linear relationships in the data.

**Mechanism:** Applies a kernel function (like polynomial, Gaussian, or RBF) to transform data into a higher-dimensional space and then performs PCA in that space.

**Applications:** Image recognition, pattern recognition, and datasets with non-linear structures.

 **Advantages**

Captures complex patterns.

Works well with non-linearly separable data.

 **Disadvantages**

Computationally expensive.

Choosing the right kernel and parameters can be challenging.

**Example Use Case**

Suppose you have a dataset of images with a non-linear relationship among the pixel intensities. Kernel PCA can help extract meaningful patterns or features that traditional PCA might miss.



# Kernal PCA Example

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import KernelPCA
from sklearn.datasets import make_circles

# Create a dataset (non-linear structure)
X, y = make_circles(n_samples=100, factor=0.3, noise=0.05, random_state=42)

# Apply Kernel PCA with RBF (Gaussian) Kernel
kpca_rbf = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca_rbf = kpca_rbf.fit_transform(X)

# Apply Kernel PCA with Polynomial Kernel
kpca_poly = KernelPCA(n_components=2, kernel='poly', degree=3, coef0=1)
X_kpca_poly = kpca_poly.fit_transform(X)

# Visualize the original dataset
plt.figure(figsize=(16, 6))

plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50, edgecolors='k')
plt.title("Original Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Visualize the transformed dataset (RBF Kernel)
plt.subplot(1, 3, 2)
plt.scatter(X_kpca_rbf[:, 0], X_kpca_rbf[:, 1], c=y, cmap='viridis', s=50, edgecolors='k')
plt.title("Kernel PCA (RBF Kernel)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

# Visualize the transformed dataset (Polynomial Kernel)
plt.subplot(1, 3, 3)
plt.scatter(X_kpca_poly[:, 0], X_kpca_poly[:, 1], c=y, cmap='viridis', s=50, edgecolors='k')
plt.title("Kernel PCA (Polynomial Kernel)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.tight_layout()
plt.show()
```
